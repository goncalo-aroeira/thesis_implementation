```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load results
tpm_df = pd.read_csv("results/elastic_net/elastic_net_results_30_pcs.csv")
fpkm_df = pd.read_csv("results/elastic_net/embeddings_30_pcs_tissue_growth_rate.csv")

# Add method labels
tpm_df["method"] = "TPM"
fpkm_df["method"] = "FPKM"

# Select relevant columns
common_cols = ["DRUG_ID", "rmse_mean", "r2_mean", "method"]
tpm_df = tpm_df[common_cols]
fpkm_df = fpkm_df[common_cols]

# Combine into one DataFrame
df = pd.concat([tpm_df, fpkm_df], ignore_index=True)

# -------- Plot RMSE Comparison -------- #
plt.figure(figsize=(10, 6))
sns.boxplot(x="rmse_mean", y="method", data=df, palette="pastel")
sns.swarmplot(x="rmse_mean", y="method", data=df, size=4, hue="method", dodge=True)

plt.title("RMSE Comparison: TPM vs FPKM (per drug)")
plt.xlabel("RMSE (Mean across CV folds)")
plt.ylabel("Gene Expression Method")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend([], [], frameon=False)
plt.tight_layout()
plt.show()

# -------- Plot R² Comparison -------- #
plt.figure(figsize=(10, 6))
sns.boxplot(x="r2_mean", y="method", data=df, palette="pastel")
sns.swarmplot(x="r2_mean", y="method", data=df, size=4, hue="method", dodge=True)

plt.title("R² Comparison: TPM vs FPKM (per drug)")
plt.xlabel("R² (Mean across CV folds)")
plt.ylabel("Gene Expression Method")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend([], [], frameon=False)
plt.tight_layout()
plt.show()
```
