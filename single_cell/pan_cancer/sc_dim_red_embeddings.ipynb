```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.manifold import TSNE
from umap import UMAP
```

```python
# Load single-cell embeddings from .npy file
sc_embeddings = np.load("embeddings/pancancer_embeddings_scfoundation.npy")

print("Shape:", sc_embeddings.shape)  # Should be (39715, 3072)
```

```python
# Reduce dimensionality to 2D with UMAP
umap_model = UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(sc_embeddings)

# Reduce dimensionality to 2D with t-SNE
tsne_model = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
X_tsne = tsne_model.fit_transform(sc_embeddings)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], s=2, alpha=0.6)
plt.title("UMAP of Single-Cell Embeddings")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], s=2, alpha=0.6)
plt.title("t-SNE of Single-Cell Embeddings")
plt.xlabel("t-SNE-1")
plt.ylabel("t-SNE-2")
plt.grid(True)
plt.tight_layout()
plt.show()
```
